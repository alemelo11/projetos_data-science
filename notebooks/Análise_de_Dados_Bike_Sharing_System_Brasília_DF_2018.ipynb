```python
import pandas as pd

# URL direta para o arquivo CSV no GitHub
url = "https://raw.githubusercontent.com/josegribeirol/bikesharingsystem_BrasiliaBR/master/data/df_rides.csv"

# Ler o arquivo CSV diretamente da URL
df = pd.read_csv(url)

# Exibir as primeiras linhas do DataFrame para verificar se o carregamento foi bem-sucedido
df.head()

```

```python

# Verificar se há valores NaN na coluna 'user_residence'
nan_values = df['user_residence'].isna().sum()
print(f"Existem {nan_values} valores NaN na coluna 'user_residence'.")

# Exibir as linhas com valores NaN na coluna 'user_residence'
print(df[df['user_residence'].isna()])

# Excluir as linhas que contêm NaN na coluna 'user_residence'
df_cleaned = df.dropna(subset=['user_residence'])

# Verificar se os valores NaN foram removidos
nan_values_after = df_cleaned['user_residence'].isna().sum()
print(f"Após a remoção, existem {nan_values_after} valores NaN na coluna 'user_residence'.")

# Exibir as primeiras linhas do DataFrame limpo
df_cleaned.head()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Converter data de nascimento para calcular idade
df['user_birthdate'] = pd.to_datetime(df['user_birthdate'])
df['user_age'] = df['user_birthdate'].apply(lambda x: datetime.now().year - x.year)

# Gráfico de Barras da Distribuição de Gênero
plt.figure(figsize=(8, 6))
sns.countplot(x='user_gender', data=df)
plt.title('Distribuição de Gênero dos Usuários')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.show()



```

```python
# Histograma da Idade dos Usuários
plt.figure(figsize=(8, 6))
sns.histplot(df['user_age'], kde=True, bins=10)
plt.title('Distribuição da Idade dos Usuários')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()

```

```python
# Supondo que a idade dos usuários já foi calculada e armazenada na coluna 'user_age'

# Calcular estatísticas descritivas
media_idade = df['user_age'].mean()
mediana_idade = df['user_age'].median()
quartis_idade = df['user_age'].quantile([0.25, 0.5, 0.75])
desvio_padrao_idade = df['user_age'].std()

# Exibir os resultados
print(f"Média da idade dos usuários: {media_idade:.2f} anos")
print(f"Mediana da idade dos usuários: {mediana_idade:.2f} anos")
print(f"Quartis da idade dos usuários:\n{quartis_idade}")
print(f"Desvio padrão da idade dos usuários: {desvio_padrao_idade:.2f} anos")
```

```python
# Boxplot da Duração das Corridas
plt.figure(figsize=(8, 6))
sns.boxplot(x='ride_duration', data=df)
plt.title('Boxplot da Duração das Corridas')
plt.xlabel('Duração (minutos)')
plt.show()

```

```python
# Gráfico de Barras para a Proporção de Corridas Atrasadas
plt.figure(figsize=(8, 6))
sns.countplot(x='ride_late', data=df, palette=['skyblue', 'orange'])
plt.title('Proporção de Corridas Atrasadas')
plt.xlabel('Corrida Atrasada')
plt.ylabel('Contagem')
plt.xticks([0, 1], ['Não', 'Sim'])
plt.show()
```
